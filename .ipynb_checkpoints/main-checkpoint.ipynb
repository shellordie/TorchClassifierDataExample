```python
from verun.utils.DatasetUtils.ClassificationDataset import ClassificationImporter
from verun.utils.DatasetUtils.ClassificationDataset import ClassificationTrainTestSpliter
from verun.utils.DatasetUtils.ClassificationDataset import ClassificationDisplayer
from verun.dataserver.core import DataServer
```

```python
server=DataServer()
datalake01="vehicle01"
trainset01="trainset01"
datapath=r"/home/charleslf/Downloads/Dataset"
server.new_datalake(datalake01)
server.new_trainset(trainset01)
#server.move_to_datalake(datapath,datalake01)
#server.move_to_trainset(datapath,trainset01)
trainset01_path=server.get_trainset_path(trainset01)
print(trainset01_path)
server.display_tree("medium")
```

```python
categories=["Bus","Car","motorcycle","Truck"]
vehi_dataset=ClassificationImporter(trainset01_path,categories,(32,32))
```

```python
trainset,testset=ClassificationTrainTestSpliter(vehi_dataset)
```

```python
ClassificationDisplayer(trainset,categories)
ClassificationDisplayer(testset,categories)
```

```python
from verun.utils.TorchUtils import ClassificationNormalizer
norm_trainset=ClassificationNormalizer(trainset)
norm_testset=ClassificationNormalizer(testset)
```

```python
import torch
batch_size=1
trainloader=torch.utils.data.DataLoader(norm_trainset,batch_size=batch_size,num_workers=2)
testloader=torch.utils.data.DataLoader(norm_testset,batch_size=batch_size,num_workers=2)
```

```python
import torch.nn as nn
import torch.nn.functional as F
class Net(nn.Module):
    def __init__(self):
        super().__init__()
        self.conv1=nn.Conv2d(3,6,5)
        self.pool1=nn.MaxPool2d(2,2)
        self.pool2=nn.MaxPool2d(2,2)
        self.conv2=nn.Conv2d(6,16,5)
        self.fc1=nn.Linear(400,120)
        self.fc2=nn.Linear(120,84)
        self.fc3=nn.Linear(84,10)
    
    def forward(self,x):
        x=self.pool1(F.relu(self.conv1(x)))
        x=self.pool2(F.relu(self.conv2(x)))
        x=torch.flatten(x,1)
        x=F.relu(self.fc1(x))
        x=F.relu(self.fc2(x))
        x=self.fc3(x)
        return x
net=Net()
print(net)
criterion=nn.CrossEntropyLoss()
optimizer=torch.optim.SGD(net.parameters(),lr=0.0001,momentum=0.9)
```

```python
for epoch in range(100):
    print("Epoch {} \n ----------------".format(epoch))
    for i,data in enumerate(trainloader,0):
        inputs,labels=data
        optimizer.zero_grad()
        outputs=net(inputs)
        loss=criterion(outputs,labels)
        loss.backward()
        optimizer.step()
        if i %10==0:
            print("loss:{}".format(loss.item()))
print("Finished Trainning")
```

```python
# TEST
```

```python
correct=0
total=0
predicted_list=[]
with torch.no_grad():
    for data in testloader:
        images,labels=data
        outputs=net(images)
        _,predicted=torch.max(outputs.data,1)
        predicted_list.append(predicted)
        total+=labels.size(0)
        correct+=(predicted==labels).sum().item()
print("accuracy on 80 test images",100*correct//total,"%")
```

```python

for i,(X,y) in enumerate(testset):
    print("ground truth",categories[y])
    print("predicted",categories[predicted_list[i]])
    print("----")
```

```python
from captum.attr import DeepLift
from captum.attr import visualization as viz
import numpy as np
import matplotlib.pyplot as plt

def attribute_image_features(algo,input,y,**kwargs):
    net.zero_grad()
    tensor_attrib=algo.attribute(input,target=y,**kwargs)
    return tensor_attrib
for i,(X,y) in enumerate(testloader):
    if i==1:
        break
    for i,(Xt,yt)in enumerate(testset):
        plt.imshow(Xt)
        plt.show()
        dl=DeepLift(net)
        attr_dl=attribute_image_features(dl,X,y,baselines=X*0)
        attr_dl=np.transpose(attr_dl.squeeze(0).cpu().detach().numpy(),(1,2,0))
        #org_img=np.transpose((Xt[i].squeeze().detach().numpy() / 2)+ 0.5, (1, 2, 0))
        _=viz.visualize_image_attr(attr_dl,Xt,method="blended_heat_map",
                                      sign="all",show_colorbar=True)
        if i==5:
            break
```
